## Data Visualization with Python, Matplotlib and Seaborn

Data visualization is graphical representation of data - producing images that communicate relationships between various aspects of the data. Visualizing data is an essential part of data analysis and machine learning

To begin, we will install the Python libraries 'Matplotlib' and 'Seaborn'. We will use the 'Matplotlib.pyplot' module for basic plots and the 'Seaborn' module will be used for advanced plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

The '%matplotlib inline' command shows the plots embedded in the jupyter notebook. Otherwise, plots open in a pop-up window occasionally

### Line Chart

A line chart is a simple and widely used technique for data viualization. This chart displays information as a series of data points connected by straight lines. All visual parameters of the chart can be modified for better visual clarity

Let's say we have a hypothetical list of rice yield across India

```python
yield_rice = [0.8, 0.85, 0.866, 0.855, 0.875, 0.92]
```

We can visualize how the yield of rice changes over time by drawing a line chart, using the 'plt.plot' function

```python
plt.plot(yield_rice)
```

The line chart is drawn as expected. We can include a ';' at the end of our statement to hide the list of plots shown in the output ([<matplotlib.lines.Line2D at 0x1c9892467c8>]

```python
plt.plot(yield_rice);
```

Let's now enhance this plot to make it more informative and appealing

### Customizing the X-Axis

The x-axis of our chart is currently showing the indexes of the list. Let's make the chart more informative by adding years to the x-axis

```python
years = [2014, 2015, 2016, 2017, 2018, 2019]
plt.plot(years, yield_rice);
```

### Adding Labels

We can add labels to show what each of the axis represent

```python
plt.plot(years, yield_rice)
plt.xlabel('Year')
plt.ylabel('Yield (tonnes per hectare)');
```

### Plotting Multiple Lines

We can plot multiple lines in the same graph. Let's compare the yield of rice to that of wheat

```python
years = range(2014, 2020)
rice = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
wheat = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908]
```

```python
plt.plot(years, rice)
plt.plot(years, wheat)
plt.xlabel('Years')
plt.ylabel('Yield (tonnes per hectare)');
```

### Chart Title and Legend

We can set a title for the chart and create a legend to differentiate between the lines in the chart

```python
plt.plot(years, rice)
plt.plot(years, wheat)

plt.xlabel('Year')
plt.ylabel('Yield (tonnes per hectare)')

plt.title("Crop Yields in India")
plt.legend(['Rice', 'Wheat']);
```

### Markers

We can show markers for our data points using the 'marker' argument of 'plt.plot'. Matplotlib provides many different markers, like a circle, cross, square, diamond etc.. A full list of markers can be found here: https://matplotlib.org/3.1.1/api/markers_api.html

```python
plt.plot(years, rice, marker='o')
plt.plot(years, wheat, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tonnes per hectare)')

plt.title("Crop Yields in India")
plt.legend(['Rice', 'Wheat']);
```

### Styling Lines and Markers

The plt.plot function supports many arguments for styling lines and markers:

    - color or c: set the color of the line
    - linestyle or ls: choose between a solid and dashed line
    - linewidth or lw: set the width of a line
    - markersize or ms: set the size of markers
    - markeredgecolor or mec: set the edge color for markers
    - markeredgewidth or mew: set the edge width for markers
    - markerfacecolor or mfc: set the fill colors for markers
    - alpha: opacity of the plot

```python
plt.plot(years, rice, marker='s', c='b', ls='-', lw=2, ms=9, mew=2, mec='navy')
plt.plot(years, wheat, marker='o', c='r', ls='--', lw=3, ms=11, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tonnes per hectare)')

plt.title("Crop Yields in India")
plt.legend(['Rice', 'Wheat']);
```
