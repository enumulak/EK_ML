#### 1. Introduction to Supervised Lerning and Regression

Supervised Learning involves training a model using labeled data with correct answers. After that, the machine is provided with a new set of data so that the supervised learning algorithm can produce the correct outcome from the labeled data with the help of training examples. Regression analysis falls under supervised learning. Here, the system tries to predict a value for an input based on previous information. The most important characteristics of regression are:

    1. The responses obtained from the model are always quantitative in nature
    2. The model can be constructed only if past data is available

Simple Linear Regression is a stastical method that allows us to summarize and study the relationship between two continuous (quantitative) variables:

    1. The first variable, usually denoted by 'x', is called the predictor, explanatory or independent variable
    2. The second variable, usually denoted by 'y', is called the response, outcome or dependent variable
    
Simple Linear Regression = Study of one predictor variable
Multiple Linear Regression = Study of multiple predictor variables

Folowing is an example of a 'deterministic' (exact) relationship between celcius and fahrenheit:

```python
celcius = [0, 10, 20, 30, 40, 50]
fahrenheit = []

for c in celcius:
    f = int((c * 9/5) + 32)
    fahrenheit.append(f)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,5), dpi=100)
ax.plot(celcius, fahrenheit, marker='o', linestyle='--');

ax.set_xlabel('Celcius')
ax.set_ylabel('Fahrenheit');
ax.set_title('Linear Relationship - Temperature');
```
